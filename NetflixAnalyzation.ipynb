Content Added Every Year

```python
# Convert date_added to datetime
data['date_added'] = pd.to_datetime(data['date_added'])

# Extract year
data['year_added'] = data['date_added'].dt.year

plt.figure(figsize=(10,6))

sns.countplot(
    data=data,
    x='year_added',
    order=sorted(data['year_added'].dropna().unique()),
    color='crimson'
)

plt.title("Content Added to Netflix Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)

plt.show()
```

 ### Observation

- The number of titles added to Netflix increased significantly after 2015, reaching a peak in the late 2010s.
- This trend highlights Netflix's rapid expansion and growing investment in content acquisition. 2015.

Top 10 Directors

```python
top_directors = (
    data['director']
    .dropna()
    .str.split(',')
    .explode()
    .str.strip()
    .value_counts()
    .head(10)
)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_directors.values,
    y=top_directors.index,
    hue=top_directors.index,
    palette='viridis'
)

plt.title("Top 10 Directors on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Director")

plt.show()
```

### Observation

- A few directors have contributed significantly more titles than others on Netflix.
- This suggests that Netflix has collaborated repeatedly with certain directors across multiple projects.

Top 10 Actors

```python
top_actors = (
    data['cast']
    .dropna()
    .str.split(',')
    .explode()
    .str.strip()
    .value_counts()
    .head(10)
)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_actors.values,
    y=top_actors.index,
    hue=top_actors.index,
    palette='magma'
)

plt.title("Top 10 Most Featured Actors on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Actor")

plt.show()
```

### Observation

- The top actors appear in multiple Netflix titles, reflecting their strong presence on the platform.
- A small group of actors contributes to a significant portion of Netflix's content library.

Monthly Release Trend

```python
data['month_added'] = data['date_added'].dt.month_name()

month_order = [
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
]

plt.figure(figsize=(12,6))

sns.countplot(
    data=data,
    x='month_added',
    order=month_order,
    color='steelblue'
)

plt.title("Monthly Content Additions on Netflix")
plt.xlabel("Month")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)

plt.show()
```

### Observation

- Content additions are spread throughout the year, with some months showing slightly higher activity than others.
- This suggests Netflix adds new titles consistently rather than concentrating releases in a specific season.
Certain months consistently receive more content additions.

Oldest vs Newest Content

```python
plt.figure(figsize=(12,6))

sns.histplot(
    data['release_year'],
    bins=30,
    color='darkred',
    kde=True
)

plt.title("Distribution of Release Years")
plt.xlabel("Release Year")
plt.ylabel("Count")

plt.show()
```

Heatmap (Numerical Correlation)

```python
# Extract duration in numeric format
data['duration_num'] = (
    data['duration']
    .str.extract(r'(\d+)')[0]
)

data['duration_num'] = pd.to_numeric(
    data['duration_num'],
    errors='coerce'
)

corr = data[['release_year','duration_num']].corr()

plt.figure(figsize=(6,4))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    linewidths=1
)

plt.title("Correlation Heatmap")

plt.show()
```

### Observation

- The heatmap indicates a **weak correlation** between release year and movie duration.
- This suggests that movie durations have remained fairly consistent over the years, with no strong linear relationship.

Top Countries by Content Rating

```python
top_countries = (
    data['country']
    .dropna()
    .str.split(',')
    .explode()
    .str.strip()
    .value_counts()
    .head(10)
    .index
)

country_rating = (
    data.assign(country=data['country'].str.split(','))
        .explode('country')
)

country_rating['country'] = country_rating['country'].str.strip()

country_rating = country_rating[
    country_rating['country'].isin(top_countries)
]

plt.figure(figsize=(14,7))

sns.countplot(
    data=country_rating,
    y='country',
    hue='rating'
)

plt.title("Content Rating Distribution Across Top Countries")
plt.xlabel("Number of Titles")
plt.ylabel("Country")

plt.show()
```

### Observation

- **TV-MA** and **TV-14** are the most common content ratings across the top Netflix-producing countries.
- The United States has the largest number of titles, with a wider variety of content ratings than other countries.
- Most countries primarily offer content targeted toward teen and adult audiences.

```python
average_movie_duration = movies.groupby('release_year')['duration_num'].mean()
std = movies.groupby('release_year')['duration_num'].std()

plt.errorbar(average_movie_duration.index, average_movie_duration.values, yerr=std.values)
plt.title('Average Movie Duration with Variation')
plt.xlabel('Year')
plt.ylabel('Minutes')
plt.show()
```

### Observation

- The average movie duration remains fairly consistent across most release years.
- The variation in movie length differs by year, indicating some years had a wider range of movie durations.
- Overall, Netflix movies generally maintain a similar runtime over time.
